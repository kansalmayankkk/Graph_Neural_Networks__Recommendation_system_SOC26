```python
import torch
import torch.nn as nn
import torch.optim as optim

# ====================================================
# DATA
# ====================================================

torch.manual_seed(42)

X = torch.linspace(
    -5,
    5,
    1000
).reshape(-1,1)

Y = X**2

# ====================================================
# MODEL
# ====================================================

class SquareNet(nn.Module):

    def __init__(self):

        super().__init__()

        self.network = nn.Sequential(

            nn.Linear(1,32),
            nn.ReLU(),

            nn.Linear(32,32),
            nn.ReLU(),

            nn.Linear(32,1)
        )

    def forward(self,x):

        return self.network(x)

model = SquareNet()

# ====================================================
# LOSS
# ====================================================

criterion = nn.MSELoss()

optimizer = optim.Adam(
    model.parameters(),
    lr=0.01
)

# ====================================================
# TRAINING
# ====================================================

epochs = 3000

for epoch in range(epochs):

    predictions = model(X)

    loss = criterion(
        predictions,
        Y
    )

    optimizer.zero_grad()

    loss.backward()

    optimizer.step()

    if (epoch+1)%300 == 0:

        print(
            f"Epoch {epoch+1} "
            f"Loss={loss.item():.8f}"
        )

# ====================================================
# TEST
# ====================================================

test_points = torch.tensor([
    [-4.0],
    [-2.0],
    [0.0],
    [2.0],
    [4.0]
])

with torch.no_grad():

    outputs = model(test_points)

print("\nPredictions")

for x,y in zip(
    test_points,
    outputs
):

    print(
        f"x={x.item():5.1f} "
        f"pred={y.item():8.4f} "
        f"true={(x.item()**2):8.4f}"
    )

# ====================================================
# PRINT WEIGHTS
# ====================================================

print("\nLearned Parameters\n")

for name,param in model.named_parameters():

    print(name)

    print(param.data)

    print()
```
