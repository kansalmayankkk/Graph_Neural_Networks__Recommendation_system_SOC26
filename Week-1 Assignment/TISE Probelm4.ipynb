```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def simulate_wavefunction(x, E, V):
    
    if E > V:
        # Oscillatory case
        k = np.sqrt(2 * (E - V))
        return np.cos(k * x)
    
    elif E < V:
        # Bound Case 
        kappa = np.sqrt(2 * (V - E))
        return np.exp(-kappa * x)
    
    else:
        # Critical Threshold (Flat-state)
        return np.ones_like(x)

if __name__ == "__main__":
    # Define the continuous spatial grid array
    # 500 points between x=0 and x=5
    x = np.linspace(0, 5, 500)
    
    # 2. Compute wavefunctions for the three distinct scenarios
    # Case A: Scattering / Oscillatory (E > V)
    psi_A = simulate_wavefunction(x, E=12, V=2)
    
    # Case B: Bound / Decaying (E < V)
    psi_B = simulate_wavefunction(x, E=2, V=7)
    
    # Case C: Threshold / Flat (E = V)
    psi_C = simulate_wavefunction(x, E=5, V=5)
    
    # 3. Generate the plot
    plt.figure(figsize=(10, 6))
    
    plt.plot(x, psi_A, label='Case A: E=12, V=2 (Oscillatory)', color='blue', linestyle='-')
    plt.plot(x, psi_B, label='Case B: E=2, V=7 (Decaying)', color='red', linestyle='--')
    plt.plot(x, psi_C, label='Case C: E=5, V=5 (Threshold)', color='green', linestyle='-.')
    
    # Formatting
    plt.title('1D Wavefunction Behavior in Constant Potential Fields', fontsize=14, fontweight='bold')
    plt.xlabel('Position $x$ (Atomic Units)', fontsize=12)
    plt.ylabel('Wavefunction $\psi(x)$', fontsize=12)
    plt.axhline(0, color='black', linewidth=0.8, alpha=0.7) # Add x-axis zero line
    plt.grid(True, alpha=0.3)
    plt.legend(loc='upper right', fontsize=10)
    
    plt.tight_layout()
    plt.show()
```
